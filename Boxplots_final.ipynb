```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.ticker as ticker

salaries = pd.read_csv('salaries.csv')
```

```python
# Define a mapping from experience_level to experience_desc
experience_mapping = {'EN': 'Entry', 'MI': 'Middle', 'SE': 'Senior', 'EX': 'Executive'}

# Create a new column 'experience_desc' using the mapping
salaries['experience_desc'] = salaries['experience_level'].map(experience_mapping)
```

```python
salaries
```

```python
# Define the order of experience levels
experience_order = ['Entry', 'Middle', 'Senior', 'Executive']

# Convert 'experience_level' to a categorical variable with the desired order
salaries['experience_level'] = pd.Categorical(salaries['experience_desc'], categories=experience_order, ordered=True)

# Set the style of seaborn for better aesthetics (optional)
sns.set(style="whitegrid")

# Create a boxplot for each experience level
plt.figure(figsize=(6,8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=salaries, palette="Set3")

#Set y-axis limits
plt.ylim(0, 500000)

# Customize the plot
plt.xlabel("")
plt.ylabel("")

# Format y-axis labels as 100k, 200k, etc.
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('${x:,.0f}'))


# Show the plot
plt.savefig('exp_boxplot.png', bbox_inches = 'tight', facecolor = 'white')
plt.show()
```
